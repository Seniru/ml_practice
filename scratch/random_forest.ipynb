```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.tree import DecisionTreeClassifier
```

```python
X, y = make_moons(n_samples=1000, noise=0.4)
```

```python
ssplit = ShuffleSplit(n_splits=1000, train_size=0.8, test_size=0.2, random_state=42)
```

```python
trees = []

for train_dex, test_dex in ssplit.split(X):
    dec_tree = DecisionTreeClassifier(max_leaf_nodes=5)
    dec_tree.fit(X[train_dex], y[train_dex])
    trees.append(dec_tree)
    print(cross_val_score(dec_tree, X[train_dex], y[train_dex], cv=3, scoring="accuracy"))
```

```python
def predict(X):
    for tree in trees:
        tree.predict(X)
    
```
