```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Getting a linear dataset

X = 2 * np.random.rand(100, 1)
y = 3 * X + 4 + np.random.rand(100, 1)
```

```python
plt.plot(X, y, "b.")
```

```python
# computing theta using the normal equation
X_b = np.c_[np.ones((100, 1)), X] # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
# making predictions from theta
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new] # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
```

```python
# plotting the regression line
plt.plot(X, y, "b.")
plt.plot(X_new, y_predict, "r-")
```

```python
# using the sklearn's LinearRegression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

lin_reg_pred = lin_reg.predict(X)

plt.plot(X, y, "b.")
plt.plot(X, lin_reg_pred, "r-")
```
