```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_moons(n_samples=1000, noise=0.2)
```

```python
df = pd.DataFrame(columns=["x", "y", "target"], data=np.c_[X, y])
plt.scatter(df["x"], df["y"], c=df["target"])
```

```python
def make_prediction_graph(min_samples_leaf):
    
    decision3 = DecisionTreeClassifier(min_samples_leaf=min_samples_leaf)
    
    decision3.fit(X, y)
    
    for x_vis in np.arange(-1.8, 2.8, 0.05):
        for y_vis in np.arange(-1.2, 1.6, 0.05):
            plt.plot(x_vis, y_vis, "b|" if decision3.predict([[x_vis, y_vis]])[0] == 0 else "r|")
        
    plt.scatter(df["x"], df["y"], c=df["target"])
    plt.show()
```

```python
make_prediction_graph(2)
```

```python
for min_leaves in range(2, 10):
    make_prediction_graph(min_leaves)
```
